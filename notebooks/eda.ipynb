```python
# Name: Raghini H | RollNo: 727823TUAM037

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import os

print("Name: Raghini H | RollNo: 727823TUAM037 | Timestamp:", datetime.now())

data = pd.read_csv("C:\\College\\Semester 6\\MLOPS\\Assignment\\ML\\data\\secom.data", sep=" ", header=None)
labels = pd.read_csv("C:\\College\\Semester 6\\MLOPS\\Assignment\\ML\\data\\secom_labels.data", sep=" ", header=None)

data['label'] = labels[0]

os.makedirs("outputs/eda", exist_ok=True)
```

```python
missing = data.isnull().sum()
plt.figure()
missing[missing > 0].plot(kind='hist')
plt.title("Missing Values Distribution")
plt.xlabel("Missing Count")
plt.ylabel("Number of Features")
plt.savefig("outputs/eda/missing_values.png")
plt.show()
plt.close()
```

```python
plt.figure()
data['label'].value_counts().plot(kind='bar')
plt.title("Class Distribution (Fault vs Normal)")
plt.xlabel("Class")
plt.ylabel("Count")
plt.savefig("outputs/eda/class_distribution.png")
plt.show()
plt.close()
```

```python
plt.figure()
data[0].hist(bins=50)
plt.title("Feature 0 Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.savefig("outputs/eda/feature_distribution.png")
plt.show()
plt.close()
```

```python
subset = data.iloc[:, :20]  
corr = subset.corr()

plt.figure()
plt.imshow(corr)
plt.colorbar()
plt.title("Correlation Heatmap (First 20 Features)")
plt.savefig("outputs/eda/correlation_heatmap.png")
plt.show()
plt.close()
```

```python
plt.figure()
data.iloc[:, :10].boxplot()
plt.title("Boxplot (First 10 Features)")
plt.savefig("outputs/eda/boxplot.png")
plt.show()
plt.close()

print("EDA completed. Plots saved in outputs/eda/")
```
